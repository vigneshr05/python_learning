```python
#dealing with bad data and converting the data from strings to integers

sales_series = pd.Series(['45.5', '20', 'N/A', '10.2', 'Error', '5'])
print(sales_series.dtype)
```

```python
# errors = 'raise' will throw an error when the bad data is found
#errors = "ignore" will stop the conversion when it finds any error
#errors = "coerce"
# Invalid values ('N/A', 'Error') are replaced with NaN
sales_series = pd.to_numeric(sales_series, errors="coerce")
print(sales_series.dtype)
```

```python
#creating a sample data file

import csv
import random
from datetime import date, timedelta

file_name = r'C:\Personal\Studies\Azure\python_learning\sample_data\sample_gen_data.csv'
num_rows = 1000

categories = ['Electronics', 'Books', 'Groceries', 'Apparel']
start_date = date(2025, 1, 1)

with open(file_name,'w', newline='') as file:
    writer = csv.writer(file,delimiter='|')
    writer.writerow(['TransactionID', 'Date', 'ProductCategory', 'Price', 'Quantity'])

    for i in range(num_rows+1):
        transaction_id = 1000 + i
        random_days = random.randint(0,364)
        transaction_date = start_date + timedelta(days=random_days)
        category = random.choice(categories)
        if random.random() < 0.05:
            price = 'VOID'
        else:
            # Generate a random float, rounded to 2 decimal places
            price = round(random.uniform(10.0, 500.0), 2)
        quantity = random.randint(1, 10)
        writer.writerow([transaction_id, transaction_date.strftime('%Y-%m-%d'), category, price, quantity])
        
```

```python
#read the above file and calculate the total revenue

custom_types = {
    'TransactionID' : 'int',
    'ProductCategory' : 'category',
    'Price' : 'float',
    'Quantity' : 'int'
}

date_cols = ["Date"]

total_revenue = 0

chunk_iterator = pd.read_csv(file_name,
                 sep="|",
                 header=0,
                 dtype= custom_types,
                 parse_dates=date_cols,
                 index_col='TransactionID',
                 chunksize=500,
                 na_values=['VOID']
                 )
```

```python
print("Processing the above file in chunks")

for i, chunk in enumerate(chunk_iterator):
    print(f"Processing chunk {i+1} with a size of {len(chunk)}")

    chunk['Price'] = pd.to_numeric(chunk['Price'],errors='coerce') #here VOID will be replaced with NaN
    chunk['Price'] = chunk['Price'].fillna(0.0)

    chunk_revenue = (chunk['Price'] * chunk['Quantity']).sum()

    total_revenue += chunk_revenue

print(f"Total Revenue: {total_revenue}")
```
